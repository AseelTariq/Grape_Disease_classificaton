# Performance Benchmarking of Modern Computer Vision Architectures

This project evaluates five distinct architectural paradigms to establish a baseline for SOTA performance:

*   Attention-based: Vision Transformer (ViT)
*   Modernized CNNs: ConvNeXt_CBAM
*   Efficient & Scalable: EfficientNet-B0
*   Hybrid Unsupervised: Autoencoder + SVM
*   Instance Segmentation: Mask R-CNN

The objective is to assess how each model handles feature extraction and generalization across the dataset, providing a holistic view of current deep learning capabilities.


## **1.Overall Performance comparison**

```python
import pandas as pd
import glob
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side

files = glob.glob("/content/*.xlsx")

input_files = [
    file for file in files
    if "Models_Comparison.xlsx" not in file
]

all_models = []

for file in input_files:
    df = pd.read_excel(file, sheet_name="Metrics")
    all_models.append(df)

comparison_df = pd.concat(all_models, ignore_index=True)

comparison_df = comparison_df[
    [
        'Model',
        'Accuracy',
        'Precision',
        'Recall',
        'F1_Score',
        'FLOPs (GMac)',
        'Train_Loss',
        'Validation_Loss',
        'Training_Time',
        'Avrage Inference_Time'
    ]
]

comparison_df['FLOPs (GMac)'] = (
    comparison_df['FLOPs (GMac)']
    .astype(str)
    .str.extract(r'([\d.]+)')
    .astype(float)
)

comparison_df.loc[
    comparison_df['Model'] == 'EfficientNet_B0',
    'FLOPs (GMac)'
] /= 1e9


comparison_df.loc[
    comparison_df['Model'].str.strip() == 'Autoencoder_SVM',
    'FLOPs (GMac)'
] *= 2

comparison_df.to_excel("Models_Comparison.xlsx", index=False)

comparison_df = comparison_df.sort_values(
    by=['Accuracy', 'F1_Score'],
    ascending=False
).reset_index(drop=True)

comparison_df.index = comparison_df.index + 1

comparison_df = comparison_df.round(4)

output_file = "Models_Comparison.xlsx"

comparison_df.to_excel(output_file, index=True)

wb = load_workbook(output_file)
ws = wb.active

header_fill = PatternFill(
    start_color="1F4E78",
    end_color="1F4E78",
    fill_type="solid"
)

header_font = Font(
    color="FFFFFF",
    bold=True
)

center_align = Alignment(
    horizontal="center",
    vertical="center"
)

thin = Side(border_style="thin", color="D9D9D9")

border = Border(
    left=thin,
    right=thin,
    top=thin,
    bottom=thin
)

for cell in ws[1]:
    cell.fill = header_fill
    cell.font = header_font
    cell.alignment = center_align
    cell.border = border

for row in ws.iter_rows():
    for cell in row:
        cell.alignment = center_align
        cell.border = border

for column in ws.columns:
    max_length = 0
    column_letter = column[0].column_letter

    for cell in column:
        try:
            if len(str(cell.value)) > max_length:
                max_length = len(str(cell.value))
        except:
            pass

    adjusted_width = max_length + 4
    ws.column_dimensions[column_letter].width = adjusted_width

for row in range(2, ws.max_row + 1):
    if row % 2 == 0:
        for cell in ws[row]:
            cell.fill = PatternFill(
                start_color="F2F2F2",
                end_color="F2F2F2",
                fill_type="solid"
            )

wb.save(output_file)

print(comparison_df.to_string())
```

## **2. Classification Quality Analysis**

# ROC Curve

```python

import pandas as pd
import matplotlib.pyplot as plt
import glob

input_files = glob.glob('/content/*_Full_Data.xlsx')

plt.figure(figsize=(10,7))

for file_path_full_data in input_files:

    model_name = (
        file_path_full_data
        .split('/')[-1]
        .replace('_Full_Data.xlsx', '')
    )

    try:

        xls = pd.ExcelFile(file_path_full_data)

        roc_sheet = None

        for sheet in xls.sheet_names:

            if (
                'roc' in sheet.lower()
                or 'auc' in sheet.lower()
            ):
                roc_sheet = sheet
                break

        if roc_sheet is None:
            print(model_name, "No ROC sheet")
            continue

        df_roc = pd.read_excel(
            file_path_full_data,
            sheet_name=roc_sheet
        )

        df_roc.columns = [
            str(col).strip().upper()
            for col in df_roc.columns
        ]

        print(model_name)
        print(df_roc.columns)

        fpr_col = None
        tpr_col = None

        for col in df_roc.columns:

            if (
                'FPR' in col
                or 'FALSE' in col
            ):
                fpr_col = col

            if (
                'TPR' in col
                or 'TRUE' in col
                or 'RECALL' in col
            ):
                tpr_col = col

        print(fpr_col, tpr_col)

        if fpr_col is None or tpr_col is None:
            continue

        plt.plot(
            df_roc[fpr_col],
            df_roc[tpr_col],
            linewidth=2.5,
            label=model_name
        )

    except Exception as e:
        print(model_name, e)

plt.plot(
    [0,1],
    [0,1],
    linestyle='--',
    color='black'
)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.title(
    "ROC Curve Comparison",
    fontsize=16,
    fontweight='bold'
)

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

# Cofusion Matrix (Best two Model)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Mapping from comparison_df model names to actual names in _Full_Data.xlsx files
# This is needed because 'PLA-ViT' in comparison_df corresponds to 'ViT' in filenames
model_name_map_to_file = {
    'PLA-ViT': 'ViT',
    'ConvNeXt_CBAM': 'ConvNeXt_CBAM',
    'Mask_R_cnn': 'Mask_R_cnn',
    'EfficientNet_B0': 'EfficientNet_B0',
    'Autoencoder_SVM': 'Autoencoder_SVM'
}

top2_models_comp_df = (
    comparison_df
    .sort_values(by='Accuracy', ascending=False)
    .head(2)['Model']
    .tolist()
)

# Get the corresponding model names as they appear in the file paths
target_file_model_names = [model_name_map_to_file.get(m, m) for m in top2_models_comp_df]

input_files_filtered = []
for f in input_files:
    # Extract the model name from the current file path
    file_model_name = f.split('/')[-1].replace('_Full_Data.xlsx', '')
    if file_model_name in target_file_model_names:
        input_files_filtered.append(f)

num_models = len(input_files_filtered)

rows = int(num_models**0.5)
cols = (num_models + rows - 1) // rows

fig, axes = plt.subplots(
    rows,
    cols,
    figsize=(cols * 7, rows * 6)
)

axes = np.array(axes).flatten()

for i, file_path_full_data in enumerate(input_files_filtered):

    model_name = (
        file_path_full_data
        .split('/')[-1]
        .replace('_Full_Data.xlsx', '')
    )

    # Determine the display name for the title based on the mapping
    display_model_name = ""
    for comp_name, file_name_part in model_name_map_to_file.items():
        if file_name_part == model_name:
            display_model_name = comp_name
            break
    if not display_model_name:
        display_model_name = model_name # Fallback

    try:

        xls = pd.ExcelFile(file_path_full_data)

        cm_sheet = None

        keywords = [
            'confusion',
            'matrix',
            'cm'
        ]

        for sheet in xls.sheet_names:

            sheet_clean = (
                sheet
                .lower()
                .replace(" ", "")
                .replace("_", "")
            )

            if any(k in sheet_clean for k in keywords):
                cm_sheet = sheet
                break

        if cm_sheet is None:
            axes[i].set_visible(False)
            continue

        df_cm = pd.read_excel(
            file_path_full_data,
            sheet_name=cm_sheet
        )

        df_cm = df_cm.set_index(df_cm.columns[0])

        ax = axes[i]

        sns.heatmap(
            df_cm,
            annot=True,
            fmt='g',
            cmap='Greens',
            linewidths=0.5,
            cbar=False,
            ax=ax
        )

        ax.set_title(
            display_model_name,
            fontsize=12,
            fontweight='bold'
        )

        ax.set_xlabel("Predicted", fontsize=10)
        ax.set_ylabel("Actual", fontsize=10)

        ax.tick_params(axis='x', labelsize=9)
        ax.tick_params(axis='y', labelsize=9)

    except Exception as e:
        print(f"{display_model_name}: Error processing Confusion Matrix: {e}")
        axes[i].set_visible(False)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle("Confusion Matrix Comparison", fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()
```

# Radar Chart

```python
import numpy as np

metrics = ['Accuracy', 'Precision', 'Recall', 'F1_Score']

labels = np.array(metrics)
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

angles = np.concatenate((angles, [angles[0]]))

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, polar=True)

for _, row in comparison_df.iterrows():

    values = row[metrics].tolist()
    values += [values[0]]

    ax.plot(angles, values, label=row['Model'])
    ax.fill(angles, values, alpha=0.1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.legend(loc='upper right')
plt.title("Model Performance Radar Chart\n\n")
plt.show()
```
